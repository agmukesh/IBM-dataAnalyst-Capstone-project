### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# Drop rows where 'DatabaseWantToWorkWith' is NaN
df_filtered = df['DatabaseWantToWorkWith'].dropna()

# Split the semicolon-separated strings and flatten them
df_series = df_filtered['DatabaseWantToWorkWith'].str.split(';').explode().str.strip()

# Count the occurrences of each database
df_count = df_series.value_counts()

# Get the top 5 most mentioned databases
top_5 = df_count.head(5)

# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(top_5, labels=top_5.index, autopct='%1.1f%%', startangle=0)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
import seaborn as sns
# Filter out missing (NaN) values
df_filtered = df['DevType'].dropna()

# Count the occurrences of each developer type
df_counts = df_filtered.value_counts()
top_5 = df_counts.head(5)

# Use Seaborn color palette
colors = sns.color_palette('pastel')

plt.figure(figsize=(5, 5))
plt.pie(top_5, labels=top_5.index, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.9)
plt.title('Top 5 Developer Roles')
plt.axis('equal')  # Keep the pie chart as a circle
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
#Drop missing values
os_series = df['OpSysProfessional use'].dropna()

# Split multiple selections (assume semicolon or comma), explode into rows, strip whitespace
os_series = os_series.str.split(';|,').explode().str.strip()

# Count the occurrences
os_counts = os_series.value_counts()

# Get the top 5 operating systems
top_os = os_counts.head(5)
labels = top_os.index
# Plot pie chart
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(top_os, labels = top_os.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
plt.title('Top Operating Systems Used Professionally')
plt.legend(wedges, labels, title="DevType", loc="center left", bbox_to_anchor=(1, 1.3))
plt.axis('equal')  # Makes the pie a circle
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
#Drop missing values from the language column
lang_series = df['LanguageHaveWorkedWith'].dropna()

# Split semicolon-separated languages and flatten into rows
lang_series = lang_series.str.split(';').explode().str.strip()

# Count frequency of each language
lang_counts = lang_series.value_counts()

# Get the top 5 languages
top_5_langs = lang_counts.head(5)

# Use a Seaborn pastel color palette
colors = sns.color_palette('pastel')

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(top_5_langs, labels=top_5_langs.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Top 5 Programming Languages Worked With')
plt.axis('equal')  # Make the pie chart a circle
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
# Filter out missing values in 'NEWCollabToolsHaveWorkedWith'
df_filtered = df[df['NEWCollabToolsHaveWorkedWith'].notna()]

# Get the top 5 collaboration tools used professionally
top_5_collab_tools = df_filtered['NEWCollabToolsHaveWorkedWith'].str.split(';').explode().value_counts().head(5)


# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_5_collab_tools, labels=top_5_collab_tools.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Collaboration Tools Used Professionally", fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a legend
plt.legend(title='Collaboration Tools', loc='upper right', bbox_to_anchor=(1.2, 1))

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
# Filter out missing values in 'LanguageAdmired'
df_filtered = df[df['LanguageAdmired'].notna()]

# Get the top 5 admired programming languages
top_5_admired_languages = (df_filtered['LanguageAdmired'].str.split(';')
                            .explode().value_counts().head(5)
                          )

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(
        top_5_admired_languages,
        labels=top_5_admired_languages.index,
        autopct='%1.1f%%',
        startangle=90
       )

plt.title("Top 5 Most Admired Programming Languages")
plt.axis('equal')

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
# Drop missing values
ai_tools = df['AIToolCurrently Using'].dropna()

# Split multiple tools per respondent (assumes semicolon delimiter), flatten into rows
ai_series = ai_tools.str.split(';').explode().str.strip()

# Count the frequency of each AI tool
ai_counts = ai_series.value_counts()

# Get the top 5 tools
top_5_ai = ai_counts.head(5)

# Use Seaborn pastel colors
colors = sns.color_palette('pastel')

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(top_5_ai, labels=top_5_ai.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Top 5 AI Tools Developers Currently Use')
plt.axis('equal')  # Make the pie chart a circle
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
# Drop missing values from the column
framework_series = df['WebframeWantToWorkWith'].dropna()

# Split semicolon-separated values, explode into rows, and clean whitespace
framework_series = framework_series.str.split(';').explode().str.strip()

# Count each framework
framework_counts = framework_series.value_counts()

# Get top 5 frameworks
top_5_frameworks = framework_counts.head(5)

# Use Seaborn color palette
colors = sns.color_palette('pastel')

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(top_5_frameworks,
        labels=top_5_frameworks.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors)
plt.title('Top 5 Web Frameworks Developers Want to Work With')
plt.axis('equal')  # Keep pie chart circular
plt.show()
```

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
# Drop NaN values
embedded_series = df['EmbeddedWantToWorkWith'].dropna()

# Split semicolon-separated values, explode into rows, and strip whitespace
embedded_series = embedded_series.str.split(';').explode().str.strip()

# Count frequency of each technology
embedded_counts = embedded_series.value_counts()

# Get the top 5 most desired embedded technologies
top_5_embedded = embedded_counts.head(5)

# Use Seaborn pastel colors
colors = sns.color_palette('pastel')

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(top_5_embedded, labels=top_5_embedded.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Top 5 Embedded Technologies Developers Want to Work With')
plt.axis('equal')  # Keep pie chart circular
plt.show()
```
