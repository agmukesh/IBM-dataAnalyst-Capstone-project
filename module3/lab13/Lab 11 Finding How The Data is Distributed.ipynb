### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
df.isnull().sum()
```

```python
df['JobSat'] = df['JobSat'].fillna(df['JobSat'].median())
```

```python
# Compute the mode (most frequent value)
remote_mode = df['RemoteWork'].mode()[0]
# Fill NaNs in the column with the mode
df['RemoteWork'] = df['RemoteWork'].fillna(remote_mode)
```

```python
#Convert to numeric, coercing errors to NaN
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# 3. Fill missing values with the median
df['YearsCodePro'] = df['YearsCodePro'].fillna(df['YearsCodePro'].median())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

```python
plt.figure(figsize=(5, 5))
colors = sns.color_palette('dark')
plt.pie(JobSat_Counts, labels=JobSat_Counts.index, startangle=140, colors=colors)
plt.title('Distribution of Job Satisfaction')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# Only keep relevant columns
df_lang = df[['LanguageHaveWorkedWith', 'LanguageWantToWorkWith']].dropna(how='all')

# Clean and explode the columns
df_lang['HaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.split(';')
df_lang['WantToWorkWith'] = df_lang['LanguageWantToWorkWith'].str.split(';')

df_have = df_lang.explode('HaveWorkedWith')['HaveWorkedWith'].str.strip()
df_want = df_lang.explode('WantToWorkWith')['WantToWorkWith'].str.strip()
```

```python
have_counts = df_have.value_counts().rename_axis('Language').reset_index(name='Used')
want_counts = df_want.value_counts().rename_axis('Language').reset_index(name='Wanted')
```

```python
lang_comparison = pd.merge(have_counts, want_counts, on='Language', how='outer').fillna(0)
lang_comparison[['Used', 'Wanted']] = lang_comparison[['Used', 'Wanted']].astype(int)

# Sort by most desired languages
lang_comparison['Gap'] = lang_comparison['Wanted'] - lang_comparison['Used']
lang_comparison.sort_values(by='Gap', ascending=False, inplace=True)
```

```python
import matplotlib.pyplot as plt

top = lang_comparison.head(15).set_index('Language')

top[['Used', 'Wanted']].plot(kind='bar', figsize=(12, 6), colormap='coolwarm', edgecolor='black')
plt.title('Most Desired vs. Used Programming Languages by gap')
plt.ylabel('Number of Respondents')
plt.xlabel('Programming Language')
plt.xticks(rotation=45)
plt.legend(title='Survey Response')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
#Step 1: Clean the data
df_filtered = df[['Country', 'RemoteWork']].dropna()

# Step 2: Compute counts
remote_counts = pd.crosstab(df_filtered['Country'], df_filtered['RemoteWork'])

# Step 3: Optional – focus on top countries
top_countries = df_filtered['Country'].value_counts().head(10).index
remote_counts = remote_counts.loc[top_countries]

# Step 4: Plot grouped bar chart
remote_counts.plot(kind='bar', figsize=(12, 6), edgecolor='black')
plt.title('Remote Work Distribution by Country')
plt.ylabel('Number of Respondents')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.legend(title='Remote Work Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Normalize to percentages
remote_percent = pd.crosstab(df_filtered['Country'], df_filtered['RemoteWork'], normalize='index') * 100
remote_percent = remote_percent.round(1).loc[top_countries]

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(remote_percent, annot=True, fmt='.1f', cmap='YlGnBu')
plt.title('Remote Work Preferences by Country (%)')
plt.xlabel('Remote Work Status')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
df_corr = df[['JobSat', 'YearsCodePro']].dropna()

# Pearson: linear correlation
pearson_corr = df_corr.corr(method='pearson').iloc[0, 1]

# Spearman: monotonic relationship (rank-based)
spearman_corr = df_corr.corr(method='spearman').iloc[0, 1]

print(f"Pearson Correlation: {pearson_corr:.3f}")
print(f"Spearman Correlation: {spearman_corr:.3f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Basic cross-tab with row percentages
edu_emp_ct = pd.crosstab(df['EdLevel'], df['Employment'], normalize='index') * 100
edu_emp_ct = edu_emp_ct.round(1)
```

```python
import matplotlib.pyplot as plt

edu_emp_ct.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set3', edgecolor='black')

plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Percentage of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
